```python
codeb = """
// OpenMP program to print Hello World
// using C language
 // OpenMP header
#include <omp.h>
#include <stdio.h>
#include <stdlib.h>
 int main(int argc, char* argv[])
{
    // Beginning of parallel region
   #pragma omp parallel
   {
       printf("Hello World... from thread = %d \\n", omp_get_thread_num());
   }
   // Ending of parallel region
}
"""
```

```python
text_file = open("code.c", "w") #keep python cells seperate from C cells
text_file.write(codeb)
text_file.close()
```

```python
%env OMP_NUM_THREADS = 3
```
